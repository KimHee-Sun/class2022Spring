```python
def test(year):
  URL = ('http://apis.data.go.kr/B552584/UlfptcaAlarmInqireSvc/'
        'getUlfptcaAlarmInfo?'
        'year='+year+
        '&pageNo=1'
        '&numOfRows=100'
        '&returnType=json'
        '&serviceKey=r2NSUyMI5q76%2B%2FrT%2B4V5KIVG%2BOMFOxlnq8og9O0fq2qhRRyPyUt6lziVVpEOW0vvxduofNAw5grn%2BGrNfHlU0Q%3D%3D')

  response = requests.get(URL)
  r = response.json()
  items = r['response']['body']['items']

  clearDate = [item['clearDate'] for item in items]
  clearTime = [item['clearTime'] for item in items]
  clearVal = [item['clearVal'] for item in items]
  dataDate = [item['dataDate'] for item in items]
  districtName = [item['districtName'] for item in items]
  issueDate = [item['issueDate'] for item in items]
  issueGbn = [item['issueGbn'] for item in items]
  issueTime = [item['issueTime'] for item in items]
  issueVal = [item['issueVal'] for item in items]
  itemCode = [item['itemCode'] for item in items]
  moveName = [item['moveName'] for item in items]
  sn = [item['sn'] for item in items]


  clearDate = pd.Series(clearDate)
  clearTime = pd.Series(clearTime)   
  clearVal = pd.Series(clearVal)
  dataDate = pd.Series(dataDate)
  districtName = pd.Series(districtName)
  issueDate = pd.Series(issueDate)
  issueGbn = pd.Series(issueGbn)
  issueTime = pd.Series(issueTime)
  issueVal = pd.Series(issueVal)
  itemCode = pd.Series(itemCode)
  moveName = pd.Series(moveName)
  sn = pd.Series(sn)

  df = pd.DataFrame({ 'clearDate': clearDate, 'clearTime': clearTime, 'clearVal': clearVal, 'dataDate': dataDate, 'districtName': districtName, 'issueDate': issueDate, 'issueGbn': issueGbn,
                   'issueTime': issueTime, 'issueVal': issueVal, 'itemCode': itemCode, 'moveName': moveName, 'sn': sn})
  return df
```

```python
iface = gr.Interface(test, gr.inputs.Dropdown(['2020']), outputs=['dataframe'])
iface.launch(debug=True)
```
